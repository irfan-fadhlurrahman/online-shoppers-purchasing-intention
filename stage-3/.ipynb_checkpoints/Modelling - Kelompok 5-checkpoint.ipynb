```python
import inflection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from matplotlib import rcParams

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from imblearn.pipeline import make_pipeline as imblearn_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, FunctionTransformer
from feature_engine.creation import CyclicalFeatures
from feature_engine.encoding import CountFrequencyEncoder, OneHotEncoder, RareLabelEncoder
from category_encoders.target_encoder import TargetEncoder

from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

rcParams['figure.figsize'] = (21, 8)
rcParams['lines.linewidth'] = 3
rcParams['xtick.labelsize'] = 'x-large'
rcParams['ytick.labelsize'] = 'x-large'
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False
rcParams['axes.spines.bottom'] = True
rcParams['axes.spines.left'] = True

SEED = 0
COLORS = ['#99d594', '#D53E4F', '#FC8D59']
pd.options.display.max_columns = 100

def load_dataset(path):
    df = pd.read_csv(path)
    df.columns = [inflection.underscore(var) for var in list(df.columns)]    
    return df

def data_preparation(df):
    df = df.copy()
    
    # drop duplicated data
    df = df.drop_duplicates()
    
    # parsing to string
    variables = ['operating_systems', 'browser', 'region', 'traffic_type', 'weekend']
    df[variables] = df[variables].astype(str)
    
    # parsing month to integer
    df['month'] = df['month'].map({
        'Feb': 2, 'Mar': 3, 'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8, 
        'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
    })
    
    # parsing revenue to integer
    df['revenue'] = df['revenue'].astype(int)
    return df

def split_dataset(df, target='revenue', test_size=0.2, seed=SEED):
    return train_test_split(
        df,
        test_size=test_size,
        random_state=seed,
        stratify=df[target]
    )

def eval_classification(model, xtrain, ytrain, xtest, ytest):
    ypred = model.predict(xtest)
    print("Accuracy (Test Set): %.2f" % accuracy_score(ytest, ypred))
    print("Precision (Test Set): %.2f" % precision_score(ytest, ypred))
    print("Recall (Test Set): %.2f" % recall_score(ytest, ypred))
    print("F1-Score (Test Set): %.2f" % f1_score(ytest, ypred))
    
    y_pred_proba = model.predict_proba(xtest)
    print("AUC: %.2f" % roc_auc_score(ytest, y_pred_proba[:, 1]))
    
def show_feature_importance(model, X_train):
    feat_importances = pd.Series(model.feature_importances_, index=X_train.columns)
    ax = feat_importances.nlargest(25).plot(kind='barh', figsize=(10, 8))
    ax.invert_yaxis()

    plt.xlabel('score')
    plt.ylabel('feature')
    plt.title('feature importance score')

def experiment():
    # data preparation
    df = load_dataset('online_shoppers_intention.csv')
    df = data_preparation(df)
    
    # set up the validation framework
    target = 'revenue'
    full_train, test = split_dataset(df, target=target, test_size=0.2)
    
    # features to use
    numerical = [
        'administrative', 'administrative_duration', 'informational',
        'informational_duration', 'product_related', 'product_related_duration',
        'bounce_rates', 'exit_rates', 'page_values', 'special_day', 
    ]
    cat_1=['month']
    cat_2=['operating_systems', 'browser', 'region', 'traffic_type']
    cat_3=['visitor_type', 'weekend']
    features = numerical + cat_1 + cat_2 + cat_3
    
    # split features and target
    X_train = full_train[features]
    y_train = full_train[target]
    X_test = test[features]
    y_test = test[target]
    
    # data preprocessing
    sin_cos_trf = CyclicalFeatures(variables=cat_1,drop_original=True)
    X_train = sin_cos_trf.fit_transform(X_train)
    X_test = sin_cos_trf.transform(X_test)
    
    target_enc = TargetEncoder(cols=cat_2)
    X_train = target_enc.fit_transform(X_train, y_train)
    X_test = target_enc.transform(X_test, y_test)
    
    one_hot_enc = OneHotEncoder(variables=cat_3)
    X_train = one_hot_enc.fit_transform(X_train)
    X_test = one_hot_enc.transform(X_test)
    
    # classifier
    model = DecisionTreeClassifier(random_state=SEED)
    model.fit(X_train, y_train)
    
    # model evaluation
    display(
        eval_classification(model, X_train, y_train, X_test, y_test),
        show_feature_importance(model, X_train)
    )
    
experiment()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import inflection

from matplotlib import rcParams
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler

from sklearn.pipeline import make_pipeline
from imblearn.pipeline import make_pipeline as imblearn_pipeline
from sklearn.compose import make_column_transformer

from sklearn.preprocessing import FunctionTransformer
from feature_engine.creation import CyclicalFeatures

from feature_engine.encoding import CountFrequencyEncoder
from feature_engine.encoding import OneHotEncoder
from feature_engine.encoding import RareLabelEncoder
from category_encoders.target_encoder import TargetEncoder

from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score

# if there is a warning to be suppresed
#import warnings
#warnings.filterwarnings("ignore")

rcParams['figure.figsize'] = (21, 8)
rcParams['lines.linewidth'] = 3
rcParams['xtick.labelsize'] = 'x-large'
rcParams['ytick.labelsize'] = 'x-large'
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False
rcParams['axes.spines.bottom'] = True
rcParams['axes.spines.left'] = True

SEED = 0
COLORS = ['#99d594', '#D53E4F', '#FC8D59']

def load_dataset(path):
    df = pd.read_csv(path)
    df.columns = [inflection.underscore(var) for var in list(df.columns)]    
    return df

def data_preparation(df):
    df = df.copy()
    
    # drop duplicated data
    df = df.drop_duplicates()
    
    # parsing to string
    variables = ['operating_systems', 'browser', 'region', 'traffic_type', 'weekend']
    df[variables] = df[variables].astype(str)
    
    # parsing month to integer
    df['month'] = df['month'].map({
        #'Jan': 1, 
        'Feb': 2, 'Mar': 3, 
        #'Apr': 4, 
        'May': 5,
        'June': 6, 'Jul': 7, 'Aug': 8, 
        'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
    })
    
    # parsing revenue to integer
    df['revenue'] = df['revenue'].astype(int)
    
    return df

def split_dataset(df, target='revenue', test_size=0.2, seed=SEED):
    return train_test_split(
        df,
        test_size=test_size,
        random_state=seed,
        stratify=df[target]
    )

def feature_transformer():
    return {
        "Log1pTransformer": FunctionTransformer(np.log1p),
        "SinCosTransformer": CyclicalFeatures()
    }

def num_scaler():
    return {
        "MinMaxScaler": MinMaxScaler(),
        "StandardScaler": StandardScaler(),
        "RobustScaler": RobustScaler(),
    }
    
def cat_encoder():
    return {
        "OneHotEncoder": OneHotEncoder(),
        "TargetEncoder": TargetEncoder(),
        "CountEncoder": CountFrequencyEncoder(encoding_method='count'),
        "FrequencyEncoder": CountFrequencyEncoder(encoding_method='frequency'),
        "RareLabelEncoder": make_pipeline(
            RareLabelEncoder(
                replace_with='rare',
                n_categories=3,
                max_n_categories=5,
            ),
            OneHotEncoder()
        ),
    }

def data_resampler(seed=SEED):
    return {
        "SMOTE": SMOTE(random_state=seed),
        "RandomUnderSampler": RandomUnderSampler(random_state=seed),
        "SMOTETomek": SMOTETomek(random_state=seed, n_jobs=-1)
    }

def linear_classifier(seed=SEED):
    clfs = {}
    clfs['LogisticRegression'] = LogisticRegression(
        random_state=seed,
        C=0.01,
        solver='lbfgs',
        class_weight='balanced',
        max_iter=5000
    )
    return clfs

def tree_based_classifier(seed=SEED):
    clfs = {}
    clfs['DecisionTreeClassifier'] = DecisionTreeClassifier(
        random_state=seed,
        max_depth=7,
        class_weight='balanced',
    )
    clfs['RandomForestClassifier'] = RandomForestClassifier(
        random_state=seed,
        max_depth=7,
        class_weight="balanced_subsample"
    )
    clfs['XGBClassifier'] = XGBClassifier(
        random_state=seed,
        n_jobs=-1,
        max_depth=7,
        scale_pos_weight=85/15
    )
    return clfs

def experiment(
        exp_id=1,
        target='revenue',
        numerical=[
                    'administrative', 'administrative_duration', 'informational',
                    'informational_duration', 'product_related', 'product_related_duration',
                    'bounce_rates', 'exit_rates', 'page_values', 'special_day', 
                  ],
        num_trf='Log1pTransformer',
        num_sclr='RobustScaler',
        cat_1=['month'], 
        cat_2=['operating_systems', 'browser', 'region', 'traffic_type'], 
        cat_3=['visitor_type', 'weekend'],
        cat_1_preprocessor='SinCosTransformer',
        cat_2_preprocessor='TargetEncoder',
        cat_3_preprocessor='OneHotEncoder',
        res_name='SMOTE',
        clf_name='LogisticRegression'
    ):
    """
    Notes:
    1. Possibility of data preprocesing per feature:
    * month (cat_1)
        sin-cos transformer, quarter binning, one-hot encoder
    * operating_systems, browser, region, traffic_type (cat_2)
        count, frequency, one-hot, target, rare-label encoder
    * visitor_type, weekend (cat_3)
        one-hot encoder
    
    2. Count and Frequency Encoder are not suitable for linear model
        
    Args:
    * exp_id, int (1): unique numbers for an experiment identifier
    * target, str ('revenue'): target variable
    * numerical, list of str : numerical features to be used
    * cat_1, cat_2, cat_3, list of str: for specifying certain columns to feed proper data preprocessing, 
    use on make_column_transformer.
    * num_trf, str: name of transformer to use, only for linear model
    * num_sclr, str: name of scaler to use, only for linear model
    * cat_1, cat_2, cat_3_preprocessor, str: name of encoder to use
    * res_name, str: name of resampler
    * clf_name, str: name of classifier
    
    Example:
    experiment(
        exp_id=1, 
        target='revenue', 
        numerical=[
                    'administrative', 'administrative_duration', 'informational',
                    'informational_duration', 'product_related', 'product_related_duration',
                    'bounce_rates', 'exit_rates', 'page_values', 'special_day', 
                  ],
        cat_1=['month'], 
        cat_2=['operating_systems', 'browser', 'region', 'traffic_type'], 
        cat_3=['visitor_type', 'weekend'],
        num_trf='Log1pTransformer',
        num_sclr='RobustScaler',
        cat_1_preprocessor='SinCosTransformer',
        cat_2_preprocessor='TargetEncoder',
        cat_3_preprocessor='OneHotEncoder',
        res_name='SMOTE',
        clf_name='LogisticRegression'
    )
    
    Return:
    * dict: 
    {
        "experiment_id": exp_id,
        "numerical_features": numerical,
        "numerical_transformer": num_trf,
        "numerical_scaler": num_sclr, 
        "categorical_features": cat_1 + cat_2 + cat_3,
        "cat_1_preprocessor":cat_1_preprocessor,
        "cat_2_preprocessor":cat_2_preprocessor,
        "cat_3_preprocessor":cat_3_preprocessor,
        "resampler_method": res_name,
        "classifier": clf_name,
        
    }
    """
    # data preparation
    df = load_dataset('online_shoppers_intention.csv')
    df = data_preparation(df)
    
    # set up the validation framework
    full_train, test = split_dataset(df, target=target, test_size=0.2)
    
    # split features and target matrices
    features = numerical + cat_1 + cat_2 + cat_3
    X_train = full_train[features]
    y_train = full_train[target]
    X_test = test[features]
    y_test = test[target]
    
    # data preprocessing and classifiers dictionaries
    transformer = feature_transformer() 
    scaler = num_scaler() 
    encoder = cat_encoder()
    resampler = data_resampler()
    classifier = linear_classifier()
    
    # preprocessor for linear model
    linear_preprocessor = make_column_transformer(
        (make_pipeline(transformer[num_trf], scaler[num_sclr]), numerical),
        (transformer[cat_1_preprocessor], cat_1),
        (encoder[cat_2_preprocessor], cat_2),
        (encoder[cat_3_preprocessor], cat_3),
        remainder='passthrough'
    )
    
    # preprocessor for tree-based model
    tree_preprocessor = make_column_transformer(
        (transformer[cat_1_preprocessor], cat_1),
        (encoder[cat_2_preprocessor], cat_2),
        (encoder[cat_3_preprocessor], cat_3),
        remainder='passthrough'
    )
    
    # pipeline for a linear algorithm
    pipe = imblearn_pipeline(
        linear_preprocessor,
        resampler[res_name],
        classifier[clf_name] # add model dictionaries
    )
    
    # fit and predict
    pipe.fit(X_train, y_train)
    
    y_pred = pipe.predict(X_train)
    clf_report = classification_report(y_train, y_pred)
    roc_auc = roc_auc_score(y_train, y_pred)
    print(f"ROC-AUC score: {roc_auc:.3f}")
    print(f"Train set classification report: \n {clf_report}")
    
    y_pred = pipe.predict(X_test)
    clf_report = classification_report(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    print(f"ROC-AUC score: {roc_auc:.3f}")
    print(f"Test set classification report: \n {clf_report}")
    
    # function result
    display(
        #full_train,
        #full_train.info(),
        #X_test,
        #y_pred.sum(),
        #linear_pipe.named_steps#['columntransfomer'].output_indices_,
        #pd.DataFrame(linear_preprocessor.fit_transform(X_train, y_train))
    )
    return {
        "experiment_id": exp_id,
        "numerical_features": numerical,
        "numerical_transformer": num_trf,
        "numerical_scaler": num_sclr, 
        "categorical_features": cat_1 + cat_2 + cat_3,
        "cat_1_preprocessor":cat_1_preprocessor,
        "cat_2_preprocessor":cat_2_preprocessor,
        "cat_3_preprocessor":cat_3_preprocessor,
        "resampler_method": res_name,
        "classifier": clf_name,
    }

trial_1 = experiment(
                        exp_id=1,
                        target='revenue',
                        numerical=[
                                    'administrative', 'administrative_duration', 'informational',
                                    'informational_duration', 'product_related', 'product_related_duration',
                                    'bounce_rates', 'exit_rates', 'page_values', 'special_day', 
                                  ],
                        num_trf='Log1pTransformer',
                        num_sclr='RobustScaler',
                        cat_1=['month'], 
                        cat_2=['operating_systems', 'browser', 'region', 'traffic_type'], 
                        cat_3=['visitor_type', 'weekend'],
                        cat_1_preprocessor='SinCosTransformer',
                        cat_2_preprocessor='TargetEncoder',
                        cat_3_preprocessor='OneHotEncoder',
                        res_name='SMOTE',
                        clf_name='LogisticRegression'
                   )

trial_1
```
